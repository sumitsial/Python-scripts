```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

### To load the .csv file containing the data values generated by the model

```python
df = pd.read_csv('Experiment_behaviourvalues.csv') # assigning the .csv file to the dataframe 'df'
```

### To print the values contained inside the .csv file

```python
print (df)
```

### The two sets of values are extracted with the help of the iloc function. 

```python
kwh = df.iloc[::2] # kwh = cost of generating per kWh of energy
co2 = df.iloc[1::2,:] # co2 = cost of abating per kg of CO2
```

### Create a boxplot of the dataset 'kwh', to examine the  cost of generating per kwh of energy for different behaviour profiles

```python
plt.figure(figsize=(10,5)) # to set figure size
sns.set() # set background for the graph
ax = sns.boxplot(y= 'values', x='attri' , hue= 'run', data=kwh) # to construct a boxplot of the 'kwh' dataset ; assigning the boxplot to a variable 'ax'
ax.set(xlabel =' ',ylabel='EUR/ kwh') # to set the labels in the x and y axis
ax.set(ylim = (0.37,0.4)) # set limits on the y-axis range
l = ax.legend() # assign the legend to a variable
l.set_title(' ') # no title for the legend 
```

### Create a boxplot of the dataset 'co2', to examine the  cost of abating per kg of co2 for different behaviour profiles

```python
plt.figure(figsize=(10,5))
sns.set()
ax = sns.boxplot(y= 'values', x='attri' , hue= 'run', data=co2) # to construct a boxplot of the 'co2' dataset
ax.set(xlabel =' ',ylabel='EUR/ kgCO2') # to set the labels in the x and y axis
ax.set(ylim = (1.280,1.350)) # set limits on the y-axis range
l = ax.legend()# assign the legend to a variable
l.set_title(' ') # no title for the legend 
```

### To load a new .csv file containing data values generated by the model

```python
df1 = pd.read_csv('Experiment_wte_behaviour.csv') # assigning the .csv file to the dataframe 'df1'
```

```python
kwh = df1.iloc[::2] # kwh = cost of generating per kWh of energy
co2 = df1.iloc[1::2,:] # co2 = cost of abating per kg of CO2
```

### Create a boxplot of the dataset 'kwh', to compare the  cost of generating per kwh of energy for different behaviour profiles and different technological routes. 

```python
plt.figure(figsize=(12,5)) 
sns.set(font_scale=1.5) # to increase the font size of the labels in the graph
ax = sns.barplot(y= 'values', x='run' , hue='route', data=kwh,palette='GnBu') # to construct a barplot of the 'kwh' dataset
ax.set(xlabel =' ',ylabel='EUR / KWh') # to set x and y axis labels
ax.set(ylim = (0.36,0.400)) # set limits on the y-axis range
l = ax.legend() # assign the legend to a variable
l.set_title(' ') # no title for the legend 
```

### Create a boxplot of the dataset 'co2', to compare the  cost of abating per kg of co2 for different behaviour profiles and different technological routes. 

```python
plt.figure(figsize=(12,5))
sns.set(font_scale=1.5)
ax = sns.barplot(y= 'values', x='run' , hue='route', data=co2,palette='GnBu')
ax.set(xlabel =' ',ylabel='EUR / kgCO2')
ax.set(ylim = (1.25,1.400))
l = ax.legend()
l.set_title(' ')
```
